# 시간 리샘플링

```python
import pandas as pd
```

```python
df = pd.read_csv('Data/starbucks.csv', index_col='Date', parse_dates=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.index
```

```python
df.resample(rule='A').mean()
```

```python
def first_day(entry):
    if len(entry):
        return entry[0]
```

```python
df.resample(rule='A').apply(first_day)
```

```python
df.head()
```

```python
df['Close'].resample('A').mean().plot.bar()
```

# Rolling과 Expanding

```python
df = pd.read_csv('Data/starbucks.csv', index_col='Date', parse_dates=True)
```

```python
df.head()
```

```python
df['Close'].plot(figsize=(12,5))
```

```python
df['Close'].plot(figsize=(12,5))
df.rolling(window=60).mean()['Close'].plot()
```

```python
df['Close: 30 Day Mean'] = df['Close'].rolling(window=30).mean()
```

```python
df.head()
```

```python
df[['Close', 'Close: 30 Day Mean']].plot(figsize=(12,5))
```

```python
df['Close'].expanding().mean().plot(figsize=(12,5))
```

# 시계열 데이터 시각화 - Part 1

```python
df = pd.read_csv('Data/starbucks.csv', index_col='Date', parse_dates=True)
```

```python
df.index
```

```python
df.plot()
```

```python
df['Close'].plot();
```

```python
df['Volume'].plot();
```

```python
title = 'TITLE'
ylabel = 'Y LABEL'
xlabel = 'X LABEL'
```

```python
ax = df['Close'].plot(figsize=(12,6) ,title=title)
ax.autoscale(axis='y', tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel)
```

```python
df['Close']['2017-01-01':'2017-12-31'].plot(figsize=(12,4))
```

```python
df['Close'].plot(figsize=(12,4), xlim=['2017-01-01','2017-12-31'],ylim=[0,70],ls='--',c='green')
```

# 시계열 데이터 시각화 - Part 2

```python
from matplotlib import dates
```

```python
ax = df['Close'].plot(xlim=['2017-01-01','2017-03-01'],ylim=[50,60], figsize=(12,5))
ax.set(xlabel='')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter("%d"))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter("\n\n%b"))

ax.yaxis.grid(True)
ax.xaxis.grid(True)
```
